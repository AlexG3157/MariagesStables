```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = pd.read_csv('temps.csv', index_col=0)
```

```python
mean = a.mean(axis=1)
```

```python
a,b,c = np.polyfit(mean.index, mean,2)
p = np.poly1d((a,b,c))
```

```python
plt.scatter(mean.index, mean, color = "navy",s=5, alpha =0.9)
plt.plot(mean.index, p(mean.index), color = "red", alpha=0.7, label=f"Fit: ${a:.2}n^2+{b:.2}n + {c:.2}$")
plt.legend()
plt.xlabel("Nombre d'étudiants ($n$)")
plt.ylabel("Temps d'exécution (secondes)")
plt.title("Temps d'exécution de l'algorithme de Gale-Shapley")
```

```python
it = pd.read_csv("iterations.csv")
mean_it = it.mean(axis=1)

a,b,c = np.polyfit(mean_it.index, mean_it,2)
p = np.poly1d((a,b,c))

plt.scatter(mean_it.index, mean_it, color = "navy",s=5, alpha =0.9)
plt.plot(mean_it.index, p(mean_it.index), color = "red", alpha=0.7, label=f"Fit: ${a:.2}n^2+{b:.2}n + {c:.2}$")
plt.legend()
plt.xlabel("Nombre d'étudiants ($n$)")
plt.ylabel("Nombre d'itérations")
plt.title("Nombre d'itérations de l'algorithme de Gale-Shapley")
```
