```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import GaleShapley
```

```python
temps, iterations = GaleShapley.mesurer_temps(200,2000,200, cote_etu=False, pire_cas=True)
```

```python
mean_temps = temps.mean(axis=1)
```

```python
a,b,c = np.polyfit(mean_temps.index, mean_temps,2)
p = np.poly1d((a,b,c))
```

```python
plt.scatter(mean_temps.index, mean_temps, color = "navy",s=5, alpha =0.9)
plt.plot(mean_temps.index, p(mean_temps.index), color = "red", alpha=0.7, label=f"Fit: ${a:.2}n^2+{b:.2}n + {c:.2}$")
plt.legend()
plt.xlabel("Nombre d'étudiants ($n$)")
plt.ylabel("Temps d'exécution (secondes)")
plt.title("Temps d'exécution de l'algorithme de Gale-Shapley")
```

```python
mean_it = iterations.mean(axis=1)

a,b,c = np.polyfit(mean_it.index, mean_it,2)
p = np.poly1d((a,b,c))

plt.scatter(mean_it.index, mean_it, color = "navy",s=5, alpha =0.9)
plt.plot(mean_it.index, p(mean_it.index), color = "red", alpha=0.7, label=f"Fit: ${a:.2}n^2+{b:.2}n + {c:.2}$")
plt.legend()
plt.xlabel("Nombre d'étudiants ($n$)")
plt.ylabel("Nombre d'itérations")
plt.title("Nombre d'itérations de l'algorithme de Gale-Shapley")
```
